# **Plotly Express 101**
## _Better than talking is actually doing. So let's code_

Biend By installing "pip install plotly-express"
Than important it with "import plotly.express as px"

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load Datframe
df = sns.load_dataset("tips")
df.head()
```

```python
px.scatter(x=[1,2,3,4,5], y=[6,7,8,9,10], title='My First Graph')
```

### **Univariate Graphics**
An univariate graphic is one that shows the distribution of just one variable of your dataset. Good examples of those types of graphics are histograms, 

boxplots, violin etc.

```python
# Histogram
fig = px.histogram(df, x='tip')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
#Boxplot
fig = px.box(df, y='tip')
fig.show()
```

```python
# Violin plot
px.violin(df, x='tip')
```

```python
# Scatter plot comparing Tips vs. Total Bill
fig = px.scatter(df, x='total_bill', y='tip', size='total_bill')
fig.show()
```

```python
# Including another variable
px.scatter(df, x='total_bill', y='tip', size='total_bill', color='sex')
```

```python
# Mean tip by size
mean_tip = df.groupby('size').tip.mean().reset_index()
px.bar(mean_tip, x='size', y='tip')
```

**3D Graphics**

Another cool feature of Plotly Express is to plot 3D graphics fairly easy

. All you gotta do is to use the function scatter_3d and add a z axis.

```python
# 3D Plot total_bill vs. tip vs. size
fig = px.scatter_3d(df, x='total_bill', y='tip', z='size', color='sex')
fig.show()
```

- The above data is really in wide range and we need to normalize and hard to compare
- Normalization change the values and to the range of 0 to 1(Now both varibale has similar)
  1. Simple Feature Scaling
     - x(new) = x(old) / x(max)
  2. Min-Max Method
  3. Z-score(standard score) -3 to +3
  4. Log Transformation

```python
# Simple feature scaling
ks4["fare"] = ks4["fare"] / ks4["fare"].max()
ks4["age in days"] = ks4["age in days"] / ks4["age in days"].max()
ks4.head()
```

```python
# Min-Max Method
ks4["fare"] = ks4["fare"] - ks4["fare"].min() / ks4["fare"].max() - ks4["fare"].min()
ks4["age in days"] = ks4["age in days"] - ks4["age in days"].min() / ks4["age in days"].max() - ks4["age in days"].min()
ks4.head()
```

```python
# Z-score (Standard Score)
ks4["fare"] = (ks4["fare"]-ks4["fare"].mean()) / ks4["fare"].std()
ks4["age in days"] = (ks4["age in days"]-ks4["age in days"].mean()) / ks4["age in days"].std()
ks4.head()
```

```python
# Log Transformation
ks4["fare"] = np.log(ks4["fare"])
ks4["age in days"] = np.log(ks4["age in days"])
ks4.head()
```

```python
# import Libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

# Load DataFrame
cancer = datasets.load_breast_cancer()
```

```python
# print the name of 30 features
print("Features:",cancer.feature_names)

# print the label type of cancer(malignent, benign)
print("Target:", cancer.target_names)
```

```python
# print data(features)shape
cancer.data.shape
```

```python
# print cancer data features(top 5 records)
print(cancer.data[0:5])
```

```python
# print the cancer labels target(1:malignent, 0:benign)
print(cancer.target)
```

```python
# import train test split function
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(cancer.data, cancer.target, test_size=0.2,random_state=0)
```

```python
# Import svm model
from sklearn import svm

# Create a svm classifier
clf = svm.SVC(kernel="linear") # Linear Kernal

# Train the model using training set 
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
y_pred
```

```python
# import scikit-Learn metrics module for accuracy calculation
from sklearn import metrics
score = metrics.accuracy_score(y_test,y_pred)
print("Accuracy score of our model is:" , (score)*100,"%")


# Model precision: What percentage of positive tuples are labled as such
score1 = metrics.precision_score(y_test,y_pred)
print("precision Score of our model is:",(score1)*100,"%")


# Model Recall: What percentage of positive tuples are labled as such
score2 = metrics.recall_score(y_test,y_pred)
print("Recall Score of our model is:",(score2)*100,"%")
```

```python
# cofusion Matrix
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```

```python
# plot confusion Matrix
sns.heatmap(cm, annot = True, cmap="Spectral")
plt.xlabel("predicted_Values")
plt.ylabel("Actual_Values")
all_sample_title = "SVM model accuracy in %:  {0}".format(score*100)
plt.title(all_sample_title, weight = "bold")
```

# **10 - Naive Bayes**

```python
# Import Libraries
import pandas as pd # ammendmets in dataframe 
import numpy as np # for arrays 
import seaborn as sns # in this tutorial this library for laoding dataframe and plotting 
import matplotlib.pyplot as plt # for plotting 
from sklearn.naive_bayes import GaussianNB # Naive Bayes library
from sklearn.model_selection import train_test_split # for splitting Dataset
from sklearn import metrics #  metrices is for checking score of our model

# load DataFrame
df = sns.load_dataset("iris")
df.head()
```

```python
# Splitting Dataframe into "X" and "y" or dependent and independent Variable
x = df.drop("species", axis = 1)
y = df["species"]
x.head()
```

```python
y.head()
```

```python
# Sppliting Dataset into Training set and test set
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)

# Create a model and fit
model = GaussianNB().fit(x_train,y_train)
print(model)

# Prediction
y_pred = model.predict(x_test)
y_pred
```

```python
# Checking Score 
score = metrics.accuracy_score(y_test,y_pred)
print("Accuracy Score of our model is",(score)*100,"%")
```

```python
# Confusion Matrix 
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```

```python
# plot confusion Matrix
sns.heatmap(cm, annot = True, cmap="Spectral")
plt.xlabel("predicted_Values")
plt.ylabel("Actual_Values")
all_sample_title = "GussianNB model accuracy in %:  {0}".format(score*100)
plt.title(all_sample_title, weight = "bold")
```
